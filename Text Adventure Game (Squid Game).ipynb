```python
#importing emoji and random package to add some creativity
import emoji
import random

######################################################################
# Constructing introduction of player and calling it to start the game
######################################################################
def player_intro():
    
    ###dynamic printing of the name to welcome the user and specifying prompt
    print('Let the squid game begin!\n')

    username = input(prompt = 'Input your name: \n')

    print(f'I would like to extend a heartfelt welcome to you, {username.title()}.\n')
    
    print(emoji.emojize("""
    You must surpass each stage to move on to the next stage. 
    If you win all games, you will receive a HUGE cash prize. :money-mouth_face::money-mouth_face::money-mouth_face:
    """))

    #declaring choice with a new prompt and converting all to lowercase
    choice = input(prompt = """You may choose from the options below to get your player number:
             A)Choose your desired number (1-456).
             B)Pick a random number.
             \n""").lower()
    
    option1 = ["choose", "your", "desired", "number"] # list of words for option A 
    option2 = ["pick", "a", "random", "number"] # list of words for option B
    choice0 = choice #variable in case the user inputs an integer
    choice = choice.split() #converting the words a user inputs into a list
    
    #using try and except if user inputs an integer
    try:
        if int(choice0):
            choice2 = int(choice0)
            
        else:
            pass
  
    except:
        choice2=choice0 #creating another variable in case the input is not an integer
            
    #developing a conditional statement for choice is "a" or any word in option1 or an integer    
    if choice0 == 'a' or any(values in option1 for values in choice) or type(choice2) == int: 
        
        #conditional statement if user inputs an integer at first
        if type(choice2) == int:
            desired_number = int(choice2)
            print(f"Got it! You are player {desired_number}!")
            stage_1()
        
        #prompting the player to input the desired number
        else:
            desired_number = input(prompt = "Enter your desired number (Hint: number between 1 and 456):")
            
            # creating a nested conditional statement 
            if int(desired_number) == 1 or int(desired_number) == 456: #conditional statement for input between 1 and 456
                print(f"Got it! You are player {desired_number}!")
                stage_1()
                
            elif int(desired_number) > 1 and int(desired_number) < 456:
                print(f"Got it! You are player {desired_number}!")
                stage_1()
                
            elif choice in option1: #conditional statement for inputting any word that is in the list option1
                print(f"Got it! You are player {desired_number}!")
                stage_1()
                
            else:
                print("Remember it must be between 1 and 456.")
                player_intro()
    
    #developing a conditional statement for choice is "b" or any word in option2
    elif choice2 == 'b' or any(values in option2 for values in choice):
        # generating a random number for the player
        rand_number = random.randint(a = 1,
                                    b = 456)
        print(f"Got it! You are player {rand_number}!")
        stage_1()
                    
    else:
        print("Your answer is not in the option.")
        player_intro() 
             
######################################################################
# Constructing stage 1: Red Light, Green Light
######################################################################
def stage_1():
    print(f"""
    {'*' * 70}
    
    Attention! Here is the first game. You will be playing, Red Light, Green Light. 
    You are allowed to move forward when â€œitâ€ shouts out, â€œGreen Light,â€. Stop when â€œitâ€ shouts, â€œRed Light.â€ 
    If you choose to move after â€œRed Lightâ€, you will be eliminated. 

    """) 
        
    #declaring action with a new prompt for green light   
    action1 = input(prompt = """â€œItâ€ shouts â€œGreen Lightâ€, what do you do? 
             1) Move forward.
             2) Stop and donâ€™t move.
             \n""").lower()
    list11 = ["move", "forward"] # list of words for option 1 
    list12 = ["stop", "and", "don't", "move"] # list of words for option 2
    action12 = action1 #variable for action that is not in the list for option 1
    action1 = action1.split() #converting the words a user inputs into a list
    
    
    #developing a conditional statement based on first action done for green light 
    if action12 == '1' or action12 == 'a' or any(answer in list11 for answer in action1):
        print("Good choice. You are a few steps ahead to the white line.\n")
              
    elif action12 == '2' or action12 == 'a' or any(answer in list12 for answer in action1):
        print("Nothing will happen to you. However, remember that you need to cross the white line before the game ends.\n")
         
    else:
        print("Invalid entry. Please try again.\nâ€")
        stage_1()
            
    #declaring action with a new prompt for green light   
    action2 = input(prompt = """â€œItâ€ shouts â€œRed Lightâ€, what do you do?
            1) Continue running or walking forward.
            2) Stop and donâ€™t move.
            \n""").lower()
    
    list21 = ["continue", "running", "or", "walking", "forward"] # list of words for option 1 
    list22 = ["stop", "and", "don't", "move"] # list of words for option 2 
    action22 = action2 #variable for action that is not in the list
    action2 = action2.split() #variable for action that is not in the list for option 2
    
    #adding another conditional statement based on second action done for red light 
    if action22 == '1' or action22 == 'a' or any(answer in list21 for answer in action2):
        fail()
        
    elif action22 == '2'  or action22 == 'b' or any(answer in list22 for answer in action2):
        print("Congratulations! You are moving on to the next stage!\n")
        stage_2()
         
    else:
        print("Invalid entry. Please try again.â€")
        stage_1()


######################################################################
# Constructing stage 2: Sugar Honeycombs
######################################################################
def stage_2():
    print(f"""
    {'*' * 70}
    
            I will now explain the rules of the next game. It is now time to make your imagination work.
            The second game is sugar honeycombs. The shape you have chosen is the shape you must remove from the honeycomb. 
            You will pass if you trim out the shape without breaking or cracking. With that, let the game begin!
    """)
    
    shape = input(prompt = "Choose one from the shapes circle,triangle,star,umbrella.\n").lower()
  
    #defining a list for shapes 
    lst_shape = ['circle', 'triangle', 'star', 'umbrella']
    shape = shape.split() #converting the words a user inputs into a list
   
    #creating a conditional statement if chosen shape is in lst_shape
    if any(answer in lst_shape for answer in shape):
        
        #creating a nested conditional statement
        if any(answer in lst_shape[0] for answer in shape) or any(answer in lst_shape[1] for answer in shape):
            print("You have a great chance of winning!\n")
             
        elif any(answer in lst_shape[2] for answer in shape) or any(answer in lst_shape[3] for answer in shape):
            print("You have to be very careful in trimming!\n")
            
        else:
            stage_2()
                
    else:
        stage_2() 
    
    #defining a variable for the result
    result = input(prompt = """How is your trimmed shape? 
             1) The shape is perfectly trimmed. 
             2) My shape broke/cracked.
             \n""").lower()
    
    outcome1 = ["the", "shape", "is", "your", "perfectly", "trimmed"] # list of words for option 1 
    outcome2 = ["my", "shape", "broke", "/", "cracked"] # list of words for option 2 
    result1 = result #variable for result that is not in the list
    result  = result.split() #splitting the words a user inputs into a list
    
    #creating a conditional statement for input as 1 or a or any word in outcome list
    if result1 == '1' or result1 == 'a' or any(ending in outcome1 for ending in result):
        print("Congratulations! You still have a chance to be a billionaire!")
        stage_3()
        
    elif result1 == '2' or result1 == 'b' or any(ending in outcome2 for ending in result):
        fail()
    
    else:
        stage_2()
    
######################################################################
# Constructing stage 3: Marble game
######################################################################
def stage_3():
    print(f"""
    {'*' * 70}
    
            For the final game, Odds and Even, you will be fighting against the only player left in the game.
            She is player 2 - Even. You are player 1 - Odd. 
            Both of you should have a set of ten marbles.
            
            In this game, you will input any number you want between 1 and 10. 
            Player 2, on the other hand, has a number in mind. If the sum of the numbers is odd, you win. 
            If even, then your opponent will win.
            
            The player who gets the best score out of three rounds gets all the marbles and the prize money. 
             
    """)
    
    #adding a list for the three rounds
    round = [1,2,3]
    
    #creating an empty list for round winner 
    round_winner = []
    
    #creating a loop for round winners
    for count in round: 
        p1_number = input(prompt = "Input any number between 1 and 10:")
        p2_number = random.randint(a = 1, b = 10)
        
        print(p1_number, p2_number)
                       
        #creating a conditional statement for numbers between 1 and 10
        if int(p1_number) >= 1 and int(p1_number) <= 10:
            
            #creating a nested conditional statement
            if (int(p1_number) + int(p2_number))%2 == 0:
                round_winner.append('player 2')
                print("You lost this round. Better luck next time.\n")
            
            else:
                round_winner.append('player 1')
                print("Yey! You won this round!\n")
        
        else:
            print("Invalid entry. Please try again.\n")
            stage_3()
            
        print(round_winner)   
     
        #creating a loop for total scores
        p1_score = 0
        p2_score = 0
        
        for score in round_winner:
            if score == 'player 1':
                p1_score = 1 + p1_score
                
            else: 
                p2_score = 1 + p2_score
                
    
#creating a conditional statement for overall winner
    if p1_score > p2_score:
            win()
                
    else:
            fail()
            
            
#defining the user-defined function fail
def fail():
    print(emoji.emojize("""You've failed. :loudly_crying_face::loudly_crying_face: Thanks for playing!"""))
    input(prompt = '<Press any key to exit>\n')    
        
def win():
    print("""
      /$$$$$$                                    /$$                    
     /$$__  $$                                  |__/                    
    | $$  \ $$ /$$$$$$/$$$$   /$$$$$$  /$$$$$$$$ /$$ /$$$$$$$   /$$$$$$ 
    | $$$$$$$$| $$_  $$_  $$ |____  $$|____ /$$/| $$| $$__  $$ /$$__  $$
    | $$__  $$| $$ \ $$ \ $$  /$$$$$$$   /$$$$/ | $$| $$  \ $$| $$  \ $$
    | $$  | $$| $$ | $$ | $$ /$$__  $$  /$$__/  | $$| $$  | $$| $$  | $$
    | $$  | $$| $$ | $$ | $$|  $$$$$$$ /$$$$$$$$| $$| $$  | $$|  $$$$$$$
    |__/  |__/|__/ |__/ |__/ \_______/|________/|__/|__/  |__/ \____  $$
                                                           /$$  \ $$
                                                          |  $$$$$$/
                                                           \______/ 
    You have finished the game. This is your ONE LUCKY DAY! You have won â‚©45,600,000,000 which is about $38,200,000.
    We all hope this is true. ðŸ˜Š
    Thank you for playing the Squid GAME!

    
    """)
    
#docstring
"""
    This game is conceptualized from the TV series, Squid Game, from Netflix. 
    The series revolves around a contest where 456 players, all of whom are in deep financial debt, risk their lives 
    to play a series of deadly children's games for the chance to win a â‚©45.6 billion prize.
    The game has three stages, Red Light Green Light, Sugar Honeycombs, and Odds and Evens. 
    To be successful, the player needs to overcome all stages.

    BUGS: For the player_intro, there will be a bug if the user inputs words and numbers together and numbers as words. 
    Also if the option is a or b in player_intro, 1 or 2 is not allowed because it will be considered the desired number.
    For all stages, the player cannot input a number which is a word, and answers with special characters (ex. "circle").
    For stage 2, the player cannot input two shapes at once.   
""" 
        
player_intro()

```
